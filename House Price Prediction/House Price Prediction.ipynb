```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

pd.pandas.set_option('display.max_columns',None)
```

```python
dataset = pd.read_csv('train.csv')

print(dataset.shape)
```

```python
dataset.head()
```

```python
features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1 ]
```

```python
features_with_na
```

```python
for feature in features_with_na:
    print(feature, np.round(dataset[feature].isnull().mean(),4),  ' %missing values')
```

```python
for feature in features_with_na:
    data = dataset.copy()
    
    data[feature] = np.where(data[feature].isnull(),1,0)
    c = ['Blue','Orange']
    
    data.groupby(feature)['SalePrice'].median().plot.bar(color=c)
    plt.title(feature)
    plt.show()
    
    
```

```python
print("The ID of Houses is {}".format(len(dataset.Id)))
```

```python
numerical_feature = [feature for feature in dataset.columns if dataset[feature].dtype != 'O']

print("The number of numerical features are {}".format(len(numerical_feature)))
```

```python
dataset[numerical_feature].head()
```

```python
Year_feature = [feature for feature in numerical_feature if 'Yr' in feature or 'Year' in feature]

Year_feature
```

```python
for feature in Year_feature:
    print(feature,data[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median of House Price')
plt.title('House price Vs Year Sold \n\n')
```

```python
for feature in Year_feature:
    if feature != 'YrSold':
        data = dataset.copy()
        data[feature] = data[feature] - data['YrSold']
        
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel('Feature')
        plt.ylabel('SalePrice')
        plt.show()
```

```python
discrete_feature=[feature for feature in numerical_feature if len(dataset[feature].unique())<25 and feature not in Year_feature+['Id']]
print("Discrete Variables Count: {}".format(len(discrete_feature)))
```

```python
discrete_feature
```

```python
for feature in discrete_feature:
    data=dataset.copy()
    cmap = plt.cm.tab10
    colors = cmap(np.arange(len(data)) % cmap.N)
    data.groupby(feature)['SalePrice'].median().plot.bar(feature,'SalePrice',color=colors)
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
continious_feature = [feature for feature in numerical_feature if feature not in discrete_feature and feature not in Year_feature + ['Id']]
print('Continious feature count : {}'.format(len(continious_feature)))
```

```python
for feature in continious_feature:
    data = dataset.copy()
    cmap = plt.cm.tab10
    colors = cmap(np.arange(len(data)) % cmap.N)
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

```python
for feature in continious_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data['Saleprice'] = np.log(data['SalePrice'])
        plt.scatter(data[feature],data['Saleprice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```

```python
## finding Outliers
for feature in continious_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
        

```
